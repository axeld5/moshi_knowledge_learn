```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
gemini_api_key = os.getenv("GEMINI_API_KEY")
```

```python
from pydantic import BaseModel

class Pair(BaseModel):
    question: str
    answer: str
```

```python
with open('files/kingdom.txt', 'r', encoding='utf-8') as file:
    content = file.read()
print(content)
```

```python
from google import genai
client = genai.Client(api_key=gemini_api_key)
```

```python
response = client.models.generate_content(
    model='gemini-2.0-flash',
    contents='Generate 100 Q/A pairs about the following text: <text>'+content+'</text> Generate the Question/Answers and only the question/answers.',
    config={
        'response_mime_type': 'application/json',
        'response_schema': list[Pair],
    },
)
# Use the response as a JSON string.
print(response.text)
```

```python
pairs = eval(response.text)
```

```python
print(len(pairs["questions"]))
print(len(pairs["answers"]))
```
