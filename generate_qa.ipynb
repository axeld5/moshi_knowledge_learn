```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
gemini_api_key = os.getenv("GEMINI_API_KEY")
```

```python
from pydantic import BaseModel

class pair(BaseModel):
    question: str
    answer: str
```

```python
with open('files/kingdom.txt', 'r', encoding='utf-8') as file:
    content = file.read()
print(content)
```

```python
from google import genai
client = genai.Client(api_key=gemini_api_key)
```

```python
response = client.models.generate_content(
    model='gemini-2.0-flash',
    contents='Generate 100 Q/A pairs about the following text: <text>'+content+'</text> Generate the Question/Answers and only the question/answers.',
    config={
        'response_mime_type': 'application/json',
        'response_schema': list[pair],
    },
)
# Use the response as a JSON string.
print(response.text)
```

```python
pairs = eval(response.text)
```

```python
import json

# Save to JSON file
with open('files/qa_pairs.json', 'w', encoding='utf-8') as f:
    json.dump(pairs, f, ensure_ascii=False, indent=2)
```

```python
new_response = client.models.generate_content(
    model='gemini-2.5-pro-exp-03-25',
    contents=f"""Given the following text: <text>{content}</text> 
    and the following QA pairs <qa_pairs>{pairs}</qa_pairs>. 
    Generate a list of 20 Q/A different pairs that can be answered if you only dispose of the information of the previous QA pairs.""",
    config={
        'response_mime_type': 'application/json',
        'response_schema': list[pair],
    },
)
# Use the response as a JSON string.
print(new_response.text)
```

```python
eval_pairs = eval(new_response.text)
```

```python
import json

# Save to JSON file
with open('files/eval_pairs.json', 'w', encoding='utf-8') as f:
    json.dump(eval_pairs, f, ensure_ascii=False, indent=2)
```
